```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.gridspec as gridspec
plt.rcParams["font.family"] = "Arial"
plt.rcParams['pdf.fonttype'] = 42
```

```python
scPA_scPAFA = pd.read_csv('/data/scPAFA_paper_formal/benchmark/plot_file/reverse/scPA_vs_scPAFA_time_mem.csv',index_col=0)
```

```python
# Creating the DataFrame
data = {
    'Category': list(scPA_scPAFA['CRC_dataset_time'].index),
    'Values': list(scPA_scPAFA['CRC_dataset_time'].values),
}
df = pd.DataFrame(data)

# Setting the color based on the Category
colors = ['#7D82B8' if 'scPAFA' not in cat else '#E5C3D1' for cat in df['Category']]

# Plotting the bar chart
plt.figure(figsize=(2, 3))
plt.bar(df['Category'], df['Values'], color=colors)
#plt.xlabel('Category')
plt.ylabel('Runtime(minute)')
plt.title('CRC dataset')
plt.xticks(rotation=90)
# Removing grid lines
plt.grid(False)
# Removing the right and top borders
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.savefig('20240806_CRC_scPA_vs_scPAFA_time.pdf',transparent=True, bbox_inches='tight')
plt.show()
```
