```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.gridspec as gridspec
plt.rcParams["font.family"] = "Arial"
plt.rcParams['pdf.fonttype'] = 42
```

```python
CRC_time = pd.read_csv('/data/scPAFA_paper_formal/benchmark/plot_file/reverse/CRC_5method_time_mem.csv',index_col=0)
```

```python
CRC_time
```

```python
Y_1 = CRC_time['fast_tl_time']
#Y_1 = Y_1/60
Y_2 = CRC_time['fast_Ucell_time']
#Y_2 = Y_2/60
Y_3 = CRC_time['tl_time']
#Y_3 = Y_3/60
Y_4 = CRC_time['Ucell_time']
#Y_4 = Y_4/60
Y_5 = CRC_time['AUcell_time']
#Y_5 = Y_5/60


plt.rcParams['font.family'] = 'Arial'
fig, ax = plt.subplots(figsize=(4, 3), dpi=300)

# --- Remove spines and add gridlines
#ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.grid(False)  # Removing all gridlines

# --- The data
# --- Fill between
#ax.fill_between(CRC_time.index, y1=Y_1, y2=Y_2, alpha=0.1,color = '#BE7C4D')

line1, = ax.plot(CRC_time.index, Y_1, marker="o", mfc="white", ms=5,color = '#53131E')
line2, = ax.plot(CRC_time.index, Y_2, marker="o", mfc="white", ms=5,color = '#BE7C4D')
line3, = ax.plot(CRC_time.index, Y_3, marker="o", mfc="white", ms=5,color = '#575C55')
line4, = ax.plot(CRC_time.index, Y_4, marker="o", mfc="white", ms=5,color = '#ACC12F')
line5, = ax.plot(CRC_time.index, Y_5, marker="o", mfc="white", ms=5,color = '#1E3888')

## --- Adjust tickers and spine to match the style of our grid
#ax.xaxis.set_major_locator(ticker.MultipleLocator(2))  # ticker every 2 matchdays
#xticks_ = ax.xaxis.set_ticklabels([x - 1 for x in range(0, len(X_) + 3, 2)])

#ax.xaxis.set_tick_params(length=2, color="#4E616C", labelcolor="#4E616C", labelsize=6)
#ax.yaxis.set_tick_params(length=2, color="#4E616C", labelcolor="#4E616C", labelsize=6)

#ax.spines["bottom"].set_edgecolor("#4E616C")

# --- Add axis labels and legend
ax.set_xlabel('Cell Number')
ax.set_ylabel('Runtime(minute)')
ax.legend(handles=[line1,line2,line3,line4,line5], loc='upper left',labels = ['fast_score_genes','fast_ucell','score_genes','UCell','AUCell'])  # Adjust location as needed

plt.suptitle('CRC dataset')

#for i, txt in enumerate(Y_1):
    #ax.annotate(txt, (pd.Series(['100k','200k','370.3k'])[i], Y_1[i]), textcoords="offset points", xytext=(0, 5), ha='center')

#for i, txt in enumerate(Y_2):
   #ax.annotate(txt, (pd.Series(['100k','200k','370.3k'])[i], Y_2[i]), textcoords="offset points", xytext=(0, 5), ha='center')
```

```python
fig.savefig('20240806_CRC_5_method_time.pdf',transparent=True, bbox_inches='tight')
```
